```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# update settings
pd.set_option('display.max_columns', None)
```

# Bar Chart Code

```python
# load data into dataframe

current_file = "obama-approval-ratings.xlsx"
obama_app_rt = pd.read_excel(current_file)
```

```python
# check the dimension of the table
print("The dimension of the table is: ", obama_app_rt.shape)
```

```python
# print the columns names
print("Column names: ", obama_app_rt.columns)
```

```python
# bar chart

obama_app_rt.plot(kind='bar',figsize=(14,8))
plt.xticks(np.arange(obama_app_rt.shape[0])[::1], obama_app_rt.Issue[::1], rotation=90)
plt.title('Obama Approval Rating', fontsize=20)
plt.suptitle('Python - Bar Chart', fontsize=15)
plt.xlabel('Issues', fontsize=12)
plt.ylabel('Approval Rating', fontsize=12)

plt.show()
```

# Stack Bar Chart Code

```python
obama_app_rt.columns
```

```python
# stack bar chart

obama_app_rt.plot(kind='bar',stacked = True, figsize=(12,8))
plt.xticks(np.arange(obama_app_rt.shape[0])[::1], obama_app_rt.Issue[::1], rotation=90)
plt.title('Obama Approval Rating', fontsize=20)
plt.suptitle('Python - Stack Bar Chart', fontsize=15)
plt.xlabel('Issues', fontsize=12)
plt.ylabel('Approval Rating', fontsize=12)

plt.show()
```

# Pie Chart Code

```python
# load data into dataframe

current_file2 = "hotdog-contest-winners.xlsx"

hotdog_contest = pd.read_excel(current_file2)
```

```python
# check the dimension of the table
print("The dimension of the table is: ", hotdog_contest.shape)
```

```python
# print the columns names
print("Column names: ", hotdog_contest.columns)
```

```python
# groupby the data by delivery type
hd_data = hotdog_contest.groupby("Country")["Winner"].count()
hd_data
```

```python
# pie chart
#Using matplotlib
pie, ax = plt.subplots(figsize=[14,10])
labels = hd_data.keys()
plt.pie(x=hd_data, autopct="%.1f%%", explode=[0.05]*4, labels=labels, pctdistance=0.5)
plt.rcParams['font.size'] = 13.0
plt.title("Hot Dog Contest Winner By Country")
plt.suptitle('Python - Pie Chart')
```

# Donut Chart Code

```python
pie, ax = plt.subplots(figsize=[14,10])
labels = hd_data.keys()
plt.pie(x=hd_data, autopct="%.1f%%", explode=[0.05]*4, labels=labels, pctdistance=0.5)
plt.rcParams['font.size'] = 13.0
plt.title("Hot Dog Contest Winner By Country")
plt.suptitle('Python - Bar Chart')
```

```python
fig1, ax = plt.subplots(figsize=(12,8))
labels = hd_data.keys()
plt.pie(x=hd_data, autopct="%.0f%%",shadow = True, labels=labels, pctdistance=0.7)
plt.rcParams['font.size']=14.0
centre_circle = plt.Circle((0,0),0.40,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Hot Dog Contest Winner By Country")
plt.suptitle('Python - Donut Chart')

ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
