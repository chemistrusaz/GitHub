```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# update settings

pd.set_option('display.max_columns', None)
```

```python
# load data

# first file
file1 = 'crimerates-by-state-2005.xlsx'
crimerates = pd.read_excel(file1)
```

```python
crimerates.columns
```

# scatterplot


```python
# plot a scatter plot

crimerates.plot.scatter(x = 'murder', y = 'burglary', s = 100)
plt.title('Python Scatter Plot: Muders against Burglary')
plt.xlabel('Amount of Muders')
plt.ylabel('Amount of Burglary')
plt.show()
```

# bubble chart

```python
# plot the bubble chart

sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.scatterplot(x='murder', 
                y='burglary',
                size='radius',
                sizes=(20,1000),
                alpha=0.5,
                data=crimerates)
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
# Put the legend out of the figure
#plt.legend(bbox_to_anchor=(1.01, 0.54),  borderaxespad=0.)
plt.xlabel('Murders')
plt.ylabel('Burglary')
plt.title('Python Bubble Chart: Murder against Burglary')
plt.tight_layout()
plt.savefig("Bubble_plot_size_range_Seaborn_scatterplot.png",
                    format='png',dpi=150)
```

# density plot chart

```python
# Density Plot and Histogram of all arrival delays

crimerates.murder.plot.density(color='green')
plt.title('Python Density plot: Muders')
plt.xlabel('Murders')
plt.show()
```
