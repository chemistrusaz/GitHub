```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import squarify    # pip install squarify (algorithm for treemap)
```

```python
# update settings

pd.set_option('display.max_columns', None)
```

```python
# load data

# first file
file1 = "expenditure_2.xlsx"
expend = pd.read_excel(file1)

# second file
file1 = "unemployement-rate-1948-2010.xlsx"
unempl = pd.read_excel(file1)

```

```python
expend.columns
```

```python
cmap = matplotlib.cm.Reds
mini=min(expend['expenditure'])
maxi=max(expend['expenditure'])
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in expend['expenditure']]
```

```python
# plot tree map

plt.figure(figsize=(10,10))
squarify.plot(sizes=expend['expenditure'], label=expend['cateogry'], color=colors, alpha=.8)
plt.axis('off')
plt.title("Python Tree Map Chart",size = 20)
plt.show()
```

# Stacked Area Chart

```python
# Your x and y axis
x=(1984,1985,1986,1987,1989)
y=[ [275,306,271,289,269], [3290,3477,3448,3664,3748], [451,529,522,562,578], [1055,1170,1149,1193,1329] ]

# stacked area chart.
plt.figure(figsize=(8,8))
plt.stackplot(x,y, labels=['Alcoholic Beverages','Food','Miscellaneous','Entertainment']) 
plt.title("Python Stacked Area Chart",size = 20)
plt.xlabel("Year", size = 12)
plt.ylabel("Amount", size = 12)
plt.legend(loc='upper left')
plt.show()
 
```

# Area Charts

```python
cat=['Alcoholic Beverages','Apparel','Cash Contributions','Education','Entertainment','Food','Healthcare','Housing',
     'Miscellaneous','Personal Care']
df = pd.DataFrame({
'Category':np.repeat(cat, 10),
'Years':list(range(2000, 2010))*10,
'Expenditure':np.random.rand(100)})
 
# Create a grid : initialize it
g = sns.FacetGrid(df, col='Category', hue='Category', col_wrap=4, )

# Add the line over the area with the plot function
g = g.map(plt.plot, 'Years', 'Expenditure')
 
# Fill the area with fill_between
g = g.map(plt.fill_between, 'Years', 'Expenditure', alpha=0.2).set_titles('{col_name} Category')
 
# Control the title of each facet
g = g.set_titles('{col_name}')
 
# Add a title for the whole plot
plt.subplots_adjust(top=0.92)
g = g.fig.suptitle('Python- Area Chart: Expenditure by Category')

# Show the graph
plt.show()
```
