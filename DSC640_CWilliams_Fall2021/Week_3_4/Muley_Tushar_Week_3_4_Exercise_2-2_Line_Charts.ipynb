```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# update settings

pd.set_option('display.max_columns', None)
```

```python
# load data

current_file = "world-population.xlsx"
wp = pd.read_excel(current_file)
```

```python
wp.head()
```

```python
# line chart

plt.figure(figsize=(10,6))
sns.lineplot(data=wp, x="Year", y="Population")
plt.title("Python Line Chart: World Population by Year",size = 20)
plt.xlabel("Year", size = 14)
plt.ylabel("Popluation", size = 14)
```

```python
# step line chart

plt.figure(figsize=(10,6))
sns.lineplot(data=wp,x="Year",y="Population", drawstyle="steps-pre", color='red', linewidth=2.5)
plt.title("Python Step Line Chart: World Population by Year",size = 20)
plt.xlabel("Year", size = 14)
plt.ylabel("Popluation", size = 14)
plt.show()
```
