```python
plt.rcParams['figure.figsize'] = (20, 10)

# make subplots
fig, axes = plt.subplots(nrows = 2, ncols = 2)

# Specify the features of interest
num_features = ['FLAG_OWN_REALTY', 'TIME_EMPLOYED', 'AGE_OF_APP', 'AMT_INCOME_TOTAL']
xaxes = num_features
yaxes = ['Counts', 'Counts', 'Counts', 'Counts']

# draw histograms
axes = axes.ravel()
for idx, ax in enumerate(axes):
    ax.hist(df_credit[num_features[idx]].dropna(), bins=40)
    ax.set_xlabel(xaxes[idx], fontsize=20)
    ax.set_ylabel(yaxes[idx], fontsize=20)
    ax.tick_params(axis='both', labelsize=15)
plt.show()
```

```python
#Lables for Legend
risky = "Risky"
not_risky = "Not Risky"
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15, 9))
#Male applicatances that are risky
men = df_credit[df_credit["CODE_GENDER"]=="M"]
#Plot
ax = sns.distplot(men[men["TARGET"]==1].AGE_OF_APP.dropna(), bins=20, label = risky, ax = axes[0], kde = False)
ax.legend()
ax.set_title('Risky Males')

#Male applicatance that are not risky
ax = sns.distplot(men[men["TARGET"]==0].AGE_OF_APP.dropna(), bins=20, label = not_risky, ax = axes[1], kde = False, color="g")
ax.legend()
_ = ax.set_title('Not Risk Males')


```

```python
#Lables for Legend
risky = "Risky"
not_risky = "Not Risky"
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15, 9))
#Women applicants that are risky
women = df_credit[df_credit["CODE_GENDER"]=="F"]
#Plot
ax = sns.distplot(women[women["TARGET"]==1].AGE_OF_APP.dropna(), bins=20, label = risky, ax = axes[0], kde =False, color="r")
ax.legend()
ax.set_title('Risky Females')

#Women applicants that are not risk
ax = sns.distplot(women[women["TARGET"]==0].AGE_OF_APP.dropna(), bins=20, label = not_risky, ax = axes[1], kde =False, color="y")
ax.legend()
_ = ax.set_title('Not Risky Females')
```

```python
#Pearson Ranking
#set up the figure size

#New features
num_features = ["FLAG_OWN_REALTY", "FLAG_OWN_CAR", "TIME_EMPLOYED", "AGE_OF_APP", "AMT_INCOME_TOTAL",
                "FLAG_MOBIL","FLAG_WORK_PHONE", "FLAG_PHONE", "FLAG_EMAIL","TARGET"]
plt.rcParams['figure.figsize'] = (15, 7)

# import the package for visulization of the correlation
from yellowbrick.features import Rank2D

# extract the numpy arrays from the data frame
X = df_credit[num_features].to_numpy()

# instantiate the visualizer with the Covariance ranking algorithm
visualizer = Rank2D(features=num_features, algorithm='pearson')
visualizer.fit(X)                # Fit the data to the visualizer
visualizer.transform(X)             # Transform the data
plt.title("Heat Map of Pearson Ranking")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()
```
