```python
#Import libraries
import pandas as pd
import numpy as np
```

For this part, you will start by reading the controversial-comments.jsonl file into a DataFrame.

```python
#Read the file
file = 'controversial-comments.jsonl' #File name
json_file = pd.read_json(file,lines=True) #Read command
json_file.head() #Check the datarame
```

```python
#Size of the dataframe
json_file.shape
```

A. Convert all text to lowercase letters.

```python
json_file["txt"] = json_file["txt"].str.lower()
#View the top 5 rows
json_file.head()
```

B. Remove all punctuation from the text.

```python
#Remove punctuation and put into a new df
json_file["txt"] = json_file["txt"].str.replace("[^\w\s]", "")
#View the top 5 rows
json_file.head()
```

C. Remove stop words.

```python
#Download stopwords 
import nltk
nltk.download('stopwords')
```

```python
#Import NLKT, stopwords and tokenize
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
```

```python
#Checking to see if the stopwords is downloaded
print(stopwords.words("english"))
```

```python
#Set the stopwords
stop = set(stopwords.words("english")) 
json_file["txt"] = json_file["txt"].apply(lambda x: [item for item in x.split() if item not in stop])

#View the data frame
json_file.head()
```

D. Apply NLTKâ€™s PorterStemmer.


```python
#Import PorterStemmer
from nltk.stem.porter import PorterStemmer
```

```python
#Create stemmer
porter = PorterStemmer()
```

```python
#Run stemmer
json_file["txt"]= json_file["txt"].apply(lambda x: [porter.stem(word) for word in x])
#View the file
json_file.head()
```
