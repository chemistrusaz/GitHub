```python
#Pull data from API for races

#My dictionary with empty lists for inputs
races = {'season': [],
        'round': [],
        'circuit_id': [],
        'lat': [],
        'long': [],
        'country': [],
        'date': [],
        'url': []}

for year in list(range(1950,2020)): #Loop for year range
    
    url = 'https://ergast.com/api/f1/{}.json' #Race URL
    r = requests.get(url.format(year)) #Request using year
    json = r.json() #Read the returning JSON

#Loop through the JSON appending data that matches list
    for item in json['MRData']['RaceTable']['Races']:
            races['season'].append(int(item['season']))
            races['round'].append(int(item['round']))
            races['circuit_id'].append(item['Circuit']['circuitId'])
            races['lat'].append(float(item['Circuit']['Location']['lat']))
            races['long'].append(float(item['Circuit']['Location']['long']))
            races['country'].append(item['Circuit']['Location']['country'])
            races['date'].append(item['date'])
            races['url'].append(item['url'])
#Take the list and make it into a dataframe
df_races = pd.DataFrame(races)
```

```python
#Loading the race_years csv to use years to pull data from API
race_years = pd.read_csv('years.csv')
```

```python
#Build rounds for year and rounds in that year

rounds = []
for year in np.array(race_years.season.unique()):
    rounds.append([year, list(race_years[race_years.season == year]['round'])])
```

```python
#Check what the rounds look like
rounds[:10]
```

```python
#Pull all the results from results dictionary within RaceTable
#Results dictionary with empty lists for data
results = {'season': [],
          'round':[],
           'circuit_id':[],
          'driver': [],
           'date_of_birth': [],
           'nationality': [],
          'constructor': [],
          'grid': [],
          'time': [],
          'status': [],
          'points': [],
          'podium': [],
          'url': []}

#Loop for through the rounds
for n in list(range(len(rounds))):
    for i in rounds[n][1]:
    
        url = 'http://ergast.com/api/f1/{}/{}/results.json' #url to call results
        r = requests.get(url.format(rounds[n][0], i)) #API request to ergast for results
        json = r.json() #Read the returning JSON

#Loop through the json to pull out the results.
        for item in json['MRData']['RaceTable']['Races'][0]['Results']: #Move through JSON to results
            results['season'].append(int(json['MRData']['RaceTable']['Races'][0]['season']))
            results['round'].append(int(json['MRData']['RaceTable']['Races'][0]['round']))
            results['circuit_id'].append(json['MRData']['RaceTable']['Races'][0]['Circuit']['circuitId'])
            results['driver'].append(item['Driver']['driverId'])
            results['date_of_birth'].append(item['Driver']['dateOfBirth'])
            results['nationality'].append(item['Driver']['nationality'])
            results['constructor'].append(item['Constructor']['constructorId'])
            results['grid'].append(int(item['grid']))
            try: #Due to missing times this assist with no returning values. Replace with NaN
                results['time'].append(int(item['Time']['millis']))
            except:#End of try block no value places NaN
                results['time'].append(None)
            results['status'].append(item['status'])
            try: #Try block due to missing points this assist with no returning values. Replace with NaN
                results['points'].append(int(item['points']))
            except:#End of try block no value places NaN
                results['points'].append(None)
            try: #Try block Due to missing podium finishers this assist no returning values. Replaces with NaN
                results['podium'].append(int(item['position']))
            except:#End of try block no value places NaN
                results['podium'].append(None)
            results['url'].append(json['MRData']['RaceTable']['Races'][0]['url'])

#Convert my dictionary to a dataframe
df_results = pd.DataFrame(results)
#Size of my dataframe
print(df_results.shape)
```
